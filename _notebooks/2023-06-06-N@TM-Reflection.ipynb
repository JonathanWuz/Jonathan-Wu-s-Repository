# The Beginning

When I first joined this course, I was nervous. I had little to no experience in computer science, and I was afraid of not being able to keep up. This fear was only heightened when my friends told me how challenging the course was. Despite this, I pushed passed my worry and registered for the class.

## Getting Adjusted

During the start of first trimester, I experienced a very new classroom experience. The room was crowded, and everyone had their computers out. There was also no paper allowed. Very peculiar. I struggled within the first week trying to install tools and dependencies and figuring out what to present. It was a very stressful start to school. But I slowly grew accustomed to the AP Computer Science Principles style, and go to work, code code coding.

## Projects

I learned a lot about working in project groups such as the importance of communication and time management. There were a lot of failures but also many successes. In terms of failure, trimester 1 is a good example. I remember our group suffered from poor time mnagement, and we were forced to code the N@tm project within the last couple of weeks. It was stressful. But successes include figuring out CRUD development and how to work with databases. Databases and CRUD was a tough hurdle for me, but through endless experimentation, I eventually got the hang of it. In the end, after each trimester, my group and I were able to produce a meaningful work at N@TM which is definetely the highlight of the year.

## Future

I'm still deciding on my future and what career I want to have but many possible careers will have a lot to do with computer science. One industry I might want to go into is engineering. Engineering is known mainly known to be physical. For example, civil engineering is the construction and planning of infastructure. But despite being very virtual and online, computer science and engineering have many parralels in the workforce. In software engineering, Computer science provides the foundation for software engineering, which involves designing, developing, and maintaining software systems. Engineers use programming languages, algorithms, and software development methodologies to create efficient and reliable software solutions for various engineering domains such as aerospace, civil, mechanical, and electrical engineering. So CS and the skills I've learned from APCSP will aid me substantially if I go into engineering. I may also become a math teacher ... not much CS involved in that career path. But if I were to become a professor or serious mathmetician, computer science is definetely important for advanced calculations.
